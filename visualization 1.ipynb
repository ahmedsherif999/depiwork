```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_csv(r"Titanic.csv")
```

```python
df
```

```python
df.head(20).plot(kind = 'bar',subplots = True)
```

```python
f = df['Fare']
f
```

```python
f.plot(kind = 'line')
plt.title('fare for each passenger')
plt.xlabel('passenger id')
plt.ylabel('fare')
```

```python
a = df['Age']
a
```

```python
a.plot(kind = 'hist')
plt.title('age of people on the titanic')
plt.xlabel('Age')
```

```python
e = df['Embarked'].value_counts()
e
```

```python
e.plot(kind = 'pie')
plt.title('embarked places')
```

```python
p = df['Pclass'].value_counts()
p
```

```python
p.plot(kind = 'bar')
```

```python
df_pclass = df.groupby('Pclass').sum()
df_pclass
```

```python
surv = df_pclass['Survived']
surv
```

```python
surv.plot(kind = 'bar')

plt.ylabel('survivals per class')
```

```python
ff = df_pclass['Fare']
ff
```

```python
ff.plot(kind = 'box')
```
